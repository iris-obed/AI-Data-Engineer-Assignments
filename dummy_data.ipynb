# **Problem 1:** Creating random numbers

```python
import numpy as np
import matplotlib.pyplot as plt

#Seed configuration for reproducibility
np.random.seed(0)

# Distribuition parameters
distribution_mean = [-3, 0]
covariance = [[1.0, 0.8], [0.8, 1.0]]

# Generate the 500 random numbers
dummy_data = np.random.multivariate_normal(distribution_mean, covariance, 500)

print("The 500 generated numbers are:")
for i, point in enumerate(dummy_data, start=1):
    print(f"{i}: {point}")


```

# **Problem 2:** Visualization by scatter plot

```python
import matplotlib.pyplot as plt

# Visualizing the data generated in Problem 1 on a scatter plot
plt.figure(figsize=(6, 6))
plt.scatter(dummy_data[:, 0], dummy_data[:, 1], color='blue', alpha=0.6, label="Random Data")
plt.title("Scatter Graph of the Data Generated")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()
plt.grid(True)
plt.show()
```

# **Problem 3**: Visualization by histogram

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(8, 8))

# HHistogram for X1
plt.subplot(2, 1, 1)
plt.hist(dummy_data[:, 0], bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.title("Histogram of x1")
plt.xlabel("x1")
plt.ylabel("Frequence")
plt.xlim(-6, 3)

# Histogram for X2
plt.subplot(2, 1, 2)
plt.hist(dummy_data[:, 1], bins=30, color='orange', alpha=0.7, edgecolor='black')
plt.title("Histogram of x2")
plt.xlabel("x2")
plt.ylabel("Frequence")
plt.xlim(-6, 3)

# Output
plt.tight_layout()
plt.show()
```

# **Problem 4:** Addition of data

```python
import numpy as np
import matplotlib.pyplot as plt

# Problem 1 data already generated
distribution_mean_1 = [1, 2]
covariance_1 = [[2.0, 0.5], [0.5, 1.0]]
dummy_data = np.random.multivariate_normal(distribution_mean_1, covariance_1, 500)

# Generating new data, problem 4
distribution_mean_4 = [0, -3]
covariance_4 = [[1.0, 0.8], [0.8, 1.0]]
new_data = np.random.multivariate_normal(distribution_mean_4, covariance_4, 500)


plt.figure(figsize=(8, 6))

# Problem 1 data
plt.scatter(dummy_data[:, 0], dummy_data[:, 1], color='blue', alpha=0.6, label='Problem 1')

# Problem 4 data
plt.scatter(new_data[:, 0], new_data[:, 1], color='orange', alpha=0.6, label='Problem 4')

# Graph configuration
plt.title("Scatter Plot: Problem 1 and Problem 4")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()
plt.grid(True)

# Output
plt.show()
```

# **Problem 5:**  Data combination

```python
import numpy as np
import pandas as pd
from google.colab import files


# Data generated in Problem 1
distribution_mean_1= [1, 2]
covariance_1 = [[2.0, 0.5], [0.5, 1.0]]
data_1 = np.random.multivariate_normal(distribution_mean_1, covariance_1, 500)

# Data generated in Problem 4
distribution_mean_4 = [0, -3]
covariance_4 = [[1.0, 0.8], [0.8, 1.0]]
data_4 = np.random.multivariate_normal(distribution_mean_4, covariance_4, 500)

# Combining the two sets of data
combined_data = np.vstack((data_1, data_4))

# Saving all records in a CSV file, as they are large
df = pd.DataFrame(combined_data, columns=["x1", "x2"])
df.to_csv("combined_data.csv", index=False)

print("Data saved in 'combined_data.csv'")

#downloading the CSV file
files.download("combined_data.csv")

# Showing the first and last 10 records of the combined data
print("The first 10 records:")
print(combined_data[:10])

print("\nThe last 10 records:")
print(combined_data[-10:])
```

# **Problem 6**: Labeling

```python
import numpy as np
import pandas as pd
from google.colab import files


# Data generated in Problem 1
distribution_mean_1 = [1, 2]
covariance_1 = [[2.0, 0.5], [0.5, 1.0]]
data_1 = np.random.multivariate_normal(distribution_mean_1, covariance_1, 500)

# Data generated in Problem 4
distribution_mean_4 = [0, -3]
covariance_4 = [[1.0, 0.8], [0.8, 1.0]]
data_4 = np.random.multivariate_normal(distribution_mean_4, covariance_4, 500)

# Combining the two sets of data
combined_data = np.vstack((data_1, data_4))

# Criating labels: 0 for data from Problem 1, 1 for data from Problem 4
labels_1 = np.zeros((data_1.shape[0], 1))
labels_4 = np.ones((data_4.shape[0], 1))

# Combining labels
combined_labels = np.vstack((labels_1, labels_4))

# Add the labels as a new column
labeled_data = np.hstack((combined_data, combined_labels))


# Saving all records in a CSV file, as they are large
df_labeled = pd.DataFrame(labeled_data, columns=["x1", "x2", "label"])
df_labeled.to_csv("labeled_data.csv", index=False)

print("Labeled data saved in 'labeled_data.csv'")

#downloading the CSV file
files.download("labeled_data.csv")

# Displaying the first 10 records as an example
print("First 10 records with labels:")
print(labeled_data[:10])


```
