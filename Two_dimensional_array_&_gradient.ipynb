# **Problem 3**: Find the gradient

```python
import numpy as np

# Creating the arrays of x and y according to Problem 1
x = np.arange(-50, 50.1, 0.1)
y = (1/2) * x + 1

# Calculating the gradient (difference between consecutive elements)
dy = np.diff(y)
dx = np.diff(x)

# Gradient at each point
gradient = dy / dx

# Showing the first 5 values as an example
print(f"The first 5 values of gradient:\n {gradient[:5]}")
print(f"Shape:\n {gradient.shape}")
```

# **Problem 4**: Draw a graph

```python
import numpy as np
import matplotlib.pyplot as plt

# Creating the arrays of x and y according to Problem 1
x = np.arange(-50, 50.1, 0.1)
y = (1/2) * x + 1

# Calculating the gradient according to Problem 3
dy = np.diff(y)
dx = np.diff(x)
gradient = dy / dx
x_gradient = x[:-1]

# Creating the graph
plt.figure(figsize=(12, 5))

# Graph of the linear function
plt.subplot(1, 2, 1)
plt.plot(x, y, label="y = (1/2)x + 1", color="blue")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graph of the Linear Function")
plt.legend()
plt.grid(True)

# Gradient graph
plt.subplot(1, 2, 2)
plt.plot(x_gradient, gradient, label="Gradient", color="red")
plt.xlabel("x")
plt.ylabel("Gradient")
plt.title("Graph of the Gradient")
plt.legend()
plt.grid(True)

# Showing graph
plt.tight_layout()
plt.show()
```

# **Problem 5**: Python functionalization

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate the gradient
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    x = np.arange(*x_range)
    y = function(x)

    # Calculating the gradient (difference between consecutive elements)
    dy = np.diff(y)
    dx = np.diff(x)
    gradient = dy / dx
    x_gradient = x[:-1]

    # Combining x and y in an array of format (n,2)
    array_xy = np.column_stack((x, y))

    return array_xy, x_gradient, gradient



# Defining mathematical functions
def function_1(x):
    return x**2

def function_2(x):
    return 2*x**2 + 2*x

def function_3(x):
    return np.sin(x**2)


# Calculating the values for each function
array_xy1, x_grad1, grad1 = compute_gradient(function_1)
array_xy2, x_grad2, grad2 = compute_gradient(function_2)
array_xy3, x_grad3, grad3 = compute_gradient(function_3, x_range=(0, 50.1, 0.1))

# Creating graphs for each function
plt.figure(figsize=(10, 6))

# Graph of the function y = x and its gradient
plt.subplot(3, 2, 1)
plt.plot(array_xy1[:, 0], array_xy1[:, 1], label="y = x", color="blue")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.subplot(3, 2, 2)
plt.plot(x_grad1, grad1, label="Gradient", color="red")
plt.xlabel("x")
plt.ylabel("Gradient")
plt.legend()
plt.grid(True)

# Graph of the function y = 2x + 2x and its gradient
plt.subplot(3, 2, 3)
plt.plot(array_xy2[:, 0], array_xy2[:, 1], label="y = 2x + 2x", color="blue")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.subplot(3, 2, 4)
plt.plot(x_grad2, grad2, label="Gradient", color="red")
plt.xlabel("x")
plt.ylabel("Gradient")
plt.legend()
plt.grid(True)

# Graph of the function y = sin(x) and its gradient
plt.subplot(3, 2, 5)
plt.plot(array_xy3[:, 0], array_xy3[:, 1], label="y = sin(x)", color="blue")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.subplot(3, 2, 6)
plt.plot(x_grad3, grad3, label="Gradient", color="red")
plt.xlabel("x")
plt.ylabel("Gradient")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

# **Problem 6**: Find the minimum value

```python
import numpy as np

# Defining mathematical functions
def function_1(x):
    return x**2

def function_2(x):
    return 2*x**2 + 2*x

def function_3(x):
    return np.sin(x**2)

# Create the function to find the minimum
def find_minimum_value(function, x_range=(-50, 50.1, 0.1)):
    x = np.arange(*x_range)
    y = function(x)

    # Find the minimum value of y and its index
    min_y = np.min(y)
    min_index = np.argmin(y)
    min_x = x[min_index]

    return min_x, min_y, min_index

# Applying the function to each equation
min_x1, min_y1, min_idx1 = find_minimum_value(function_1)
min_x2, min_y2, min_idx2 = find_minimum_value(function_2)
min_x3, min_y3, min_idx3 = find_minimum_value(function_3, x_range=(0, 50.1, 0.1))

# Showing the results
print(f"For y = x, the minimum occurs at x = {min_x1}, y = {min_y1}")
print(f"For y = 2x + 2x, the minimum occurs at x = {min_x2}, y = {min_y2}")
print(f"For y = sin(x), the minimum occurs at x = {min_x3}, y = {min_y3}")
```
