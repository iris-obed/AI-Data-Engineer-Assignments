```python
import math
import matplotlib.pyplot as plt

# Constants
# Volume of a single chestnut bun (approximation in cubic meters)
volume_chestnut_bun = 0.0005

# Volume of the solar system (approximation in cubic meters)
# Assuming the solar system radius as 100 AU (1 AU = 1.496 x 10^11 meters)
radius_solar_system_m = 100 * 1.496e11
volume_solar_system = (4 / 3) * math.pi * radius_solar_system_m**3

# Time interval for doubling (5 minutes)
time_interval_minutes = 5

# Generic function to calculate time to fill a space with "Byvine"
def time_to_fill_space(initial_volume, target_volume, doubling_time):

    # Returns the total time required in minutes

    current_volume = initial_volume
    time_elapsed = 0

    while current_volume < target_volume:
        current_volume *= 2
        time_elapsed += doubling_time

    return time_elapsed

# Calculate the time to fill the solar system with chestnut buns
time_to_fill_solar_system = time_to_fill_space(
    volume_chestnut_bun, volume_solar_system, time_interval_minutes
)

# Funtion to generate a graph for visualization
def generate_growth_graph(initial_volume, target_volume, doubling_time):

    current_volume = initial_volume
    time_elapsed = 0
    times = [time_elapsed]
    volumes = [current_volume]

    while current_volume < target_volume:
        current_volume *= 2
        time_elapsed += doubling_time
        times.append(time_elapsed)
        volumes.append(current_volume)

    plt.figure(figsize=(8, 6))
    plt.plot(times, volumes, label="Volume Growth", color="orange")
    plt.axhline(y=target_volume, color="green", linestyle="--", label="Target Volume")
    plt.title("Growth of Chestnut Buns Over Time", color="blue")
    plt.xlabel("Time (minutes)", color="gray")
    plt.ylabel("Volume (mÂ³)", color="gray")
    plt.legend()
    plt.grid(True)
    plt.show()

# Results
time_in_days = time_to_fill_solar_system / (24 * 60)  # Convert minutes to days
print(f"Time to fill the solar system with chestnut buns: {time_in_days:.2e} days")

# Generate the growth graph
generate_growth_graph(volume_chestnut_bun, volume_solar_system, time_interval_minutes)
```
