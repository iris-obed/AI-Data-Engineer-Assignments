#**Problem 4**: Dealing with missing values

```python
import pandas as pd
import missingno as msno

# Load the dataset
df = pd.read_csv("train.csv")

# Check for missing values
missing_values = df.isnull().sum()
missing_ratio = (missing_values / len(df)).sort_values(ascending=False)

# Create a DataFrame with missing value statistics
missing_data = pd.DataFrame({
    "Total Missing": missing_values,
    "Missing Ratio": missing_ratio
})

# Display only features with missing values
missing_data = missing_data[missing_data["Total Missing"] > 0]
print("Missing values per feature:\n")
print(missing_data, "\n")

# Visualize missing data using missingno
msno.matrix(df)

# Remove features (columns) with 5 or more missing values
threshold = 5
df_cleaned = df.drop(columns=missing_data[missing_data["Total Missing"] >= threshold].index)
print(f"Removed {len(missing_data[missing_data['Total Missing'] >= threshold])} features with 5 or more missing values.\n")

# Remove rows with any remaining missing values
df_cleaned = df_cleaned.dropna()
print(f"Remaining dataset shape after removing missing data: {df_cleaned.shape}")
```

#**Question 5**: Terminology research

#**1. Kurtosis**

**Definition**:

Kurtosis measures the “tailedness” of a probability distribution. It indicates whether the data have heavy or light tails compared to a normal distribution.
* **High kurtosis (>3)**: The distribution has heavy tails, meaning more outliers.
*	**Low kurtosis (<3)**: The distribution has light tails, meaning fewer extreme values.
*	**Normal kurtosis (≈3)**: Matches a normal distribution (bell-shaped curve).

**Example**:
*	A dataset with frequent extreme values will have high kurtosis.
*	A uniform distribution has low kurtosis.



#**2. Skewness**

**Definition**:

Skewness measures the asymmetry of a distribution. It tells us whether the data are concentrated more on one side of the mean.
*	**Positive skew (Right-skewed, >0)**: The right tail is longer, meaning more high-value outliers. Example: Income distribution.
*	**Negative skew (Left-skewed, <0)**: The left tail is longer, meaning more low-value outliers. Example: Test scores where most students score high.
*	**Zero skewness (≈0)**: The distribution is symmetrical.

**Example**:
*	If house prices have positive skewness, most houses are cheap, but a few very expensive houses pull the average up.




#**Problem 6**: Checking the distribution

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import kurtosis, skew
import numpy as np

# Load the dataset
df = pd.read_csv("train.csv")

# Extract the target variable (SalePrice)
target_var = "SalePrice"

# Calculate kurtosis and skewness before transformation
original_kurtosis = kurtosis(df[target_var])
original_skewness = skew(df[target_var])

print(f"Original Kurtosis: {original_kurtosis:.2f}")
print(f"Original Skewness: {original_skewness:.2f}")

# Plot the original distribution
plt.figure(figsize=(12, 5))
sns.histplot(df[target_var], kde=True, bins=30)
plt.title("Original SalePrice Distribution")
plt.xlabel("SalePrice")
plt.ylabel("Frequency")
plt.show()

# Apply log transformation
df["Log_SalePrice"] = np.log1p(df[target_var])

# Calculate kurtosis and skewness after transformation
log_kurtosis = kurtosis(df["Log_SalePrice"])
log_skewness = skew(df["Log_SalePrice"])

print(f"Log-Transformed Kurtosis: {log_kurtosis:.2f}")
print(f"Log-Transformed Skewness: {log_skewness:.2f}")

# Plot the log-transformed distribution
plt.figure(figsize=(12, 5))
sns.histplot(df["Log_SalePrice"], kde=True, bins=30)
plt.title("Log-Transformed SalePrice Distribution")
plt.xlabel("Log(SalePrice)")
plt.ylabel("Frequency")
plt.show()
```

#**Question 7**: Checking correlation coefficients

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
df = pd.read_csv("train.csv")

# Check if there are non-numeric columns and remove them for correlation calculation
numeric_df = df.select_dtypes(include=[np.number])

# Compute the correlation matrix
correlation_matrix = numeric_df.corr()

# Plot heatmap for all features
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap for All Features")
plt.show()

# Select the 10 most correlated features with SalePrice
top_10_features = correlation_matrix["SalePrice"].abs().sort_values(ascending=False)[1:11].index

# Create a heatmap for the selected 10 features
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df[top_10_features].corr(), cmap="coolwarm", annot=True, fmt=".2f")
plt.title("Top 10 Features Correlated with SalePrice")
plt.show()

# Find highly correlated feature pairs
high_corr_pairs = []
threshold = 0.7  # Define high correlation threshold

for i in range(len(top_10_features)):
    for j in range(i+1, len(top_10_features)):
        corr_value = numeric_df[top_10_features[i]].corr(numeric_df[top_10_features[j]])
        if abs(corr_value) > threshold:
            high_corr_pairs.append((top_10_features[i], top_10_features[j], corr_value))

# Display high correlation feature pairs
print("Highly Correlated Feature Pairs (Above 0.7):")
for pair in high_corr_pairs:
    print(f"{pair[0]} and {pair[1]} → Correlation: {pair[2]:.2f}")
```
