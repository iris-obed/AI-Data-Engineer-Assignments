```python
# Problem 1: Number of grains of rice on the 100th day

import matplotlib.pyplot as plt

# Variable initialization
days = 100
grains_per_day = []
total_rice = 0

# Number of grains of rice for each day and the total
for day in range(days):
    rice_today = 2 ** day
    grains_per_day.append(rice_today)
    total_rice += rice_today

# Displaying the results
rice_on_100th_day = grains_per_day[-1]
print(f"Number of grains of rice on the 100th day: {rice_on_100th_day}")
print(f"Total amount of rice received by day 100: {total_rice}")

# Creating and customizing a line graph

plt.figure(figsize=(10, 6))
plt.plot(range(1, days + 1), grains_per_day, label='Grains of Rice on Each Day', color='orange')
plt.plot(range(1, days + 1), [sum(grains_per_day[:d]) for d in range(1, days + 1)], label='Total Grains of Rice', color='green')
plt.title('Rice Grains Received by Day', color='Blue')
plt.xlabel('Days',color='gray')
plt.ylabel('Number of Grains of Rice' , color='gray')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Problem 2: Number of rice grains outside of the 100th day

import matplotlib.pyplot as plt

def compute_sorori_shinzaemon(n_days):

    list_n_grains = [2**(day - 1) for day in range(1, n_days + 1)]
    list_total_grains = [sum(list_n_grains[:day]) for day in range(1, n_days + 1)]
    return list_n_grains, list_total_grains

# Request number of days from the user
n_days = int(input("Enter the numbers of days: "))
if n_days <1:
    print("Invalid number! Please, enter a number higher than zero")
else:
    # Calculating the grains of rice
    list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days)

    # Graphic config
    plt.plot(range(1, n_days + 1), list_n_grains, label="Grains of Rice on Each Day", color="cyan")
    plt.plot(range(1, n_days + 1), list_total_grains, label="Total Grains of Rice", color="orange")
    plt.title(f"Rice grains in {n_days} days"  , color='blue')
    plt.xlabel("Days"  , color='gray')
    plt.ylabel("Number of Grains"  , color='gray')
    plt.legend()
    plt.grid()
    plt.show()

    # Results
    print(f"Grains of rice on {n_days} day: {list_n_grains[-1]}")
    print(f"Total grains of rice on {n_days} day: {list_total_grains[-1]}")
```

```python
#Problem 3: How many people can live for how many days with the rice we get

""" According to the website https://en.wikipedia.org/wiki/Freerice,
the average adult consumes approximately 400 grams of rice per day,
which is equivalent to approximately 19,200 grains of rice. """

def calculate_days_to_live(total_rice, num_people):

    #Calculates the number of days a class can live on a specific amount of rice
    rice_per_person_per_day = 19200
    return total_rice // (num_people * rice_per_person_per_day)


# Total accumulated grains in 100 days
total_rice = 2**100 - 1

# Entering the number of people in the class
num_people = int(input("Enter the number of people in class: "))

if num_people > 0:
    days_to_live = calculate_days_to_live(total_rice, num_people)
    print(f"A class of {num_people} people can live for {days_to_live} days with {total_rice} grains of rice.")
else:
    print("There has to be at least one person in the class.")
```
