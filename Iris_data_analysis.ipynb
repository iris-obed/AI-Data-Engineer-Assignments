#**Problem 1**: Acquiring data

```python
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()

# Create a DataFrame for the explanatory variables (features)
X = pd.DataFrame(iris.data, columns=iris.feature_names)

# Create a DataFrame for the target variable
y = pd.DataFrame(iris.target, columns=['Species'])

# Display the first rows of the DataFrames
print("Explanatory Variables (X):")
print(X.head())

print("\nTarget Variable (y):")
print(y.head())
```

```python
# Combining X and Y in a single DataFrame
df_combined = pd.concat([X, y], axis=1)

# Renaming the columns to match the example provided
df_combined.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

# Displaying the first rows of the combined DataFrame
print("DataFrame combinado:")
print(df_combined.head())
```

#**Problem 5**: Extracting necessary data

```python
# Complete Solution for Problem 5: Data Extraction from Iris DataFrame

import pandas as pd
from sklearn.datasets import load_iris

# Load Iris dataset and create clean DataFrame
iris = load_iris()
df = pd.DataFrame(
    iris.data,
    columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
)
df['species'] = iris.target

print("="*80)
print("Complete Iris DataFrame:")
print(df.head())

# Extract column using two different methods
print("\n" + "="*80)
print("Extracting sepal_width column:")
print("\na) Recommended method (bracket notation df['column']):")
print(df['sepal_width'].head())

print("\nb) Dot notation method (df.column):")
print(df.sepal_width.head())

# Extract rows 50 to 99 (indices 49 to 98)
print("\n" + "="*80)
print("Rows 50 to 99 (indices 49 to 98):")
print(df.iloc[49:99].head())

# Extract petal_length only for rows 50-99
print("\n" + "="*80)
print("Petal_length values for rows 50-99:")
print(df.iloc[49:99]['petal_length'].head())

# Filter rows where petal_width equals 0.2
print("\n" + "="*80)
print("Rows with petal_width = 0.2:")
print(df[df['petal_width'] == 0.2])

# Demonstrate differences between .loc and .iloc
print("\n" + "="*80)
print("Key differences between .loc and .iloc:")

print("\na) .loc - Label-based selection (inclusive):")
print("df.loc[49:99] includes row 99:")
print(df.loc[49:99].tail(1))

print("\nb) .iloc - Position-based selection (exclusive):")
print("df.iloc[49:99] excludes row 99:")
print("Last returned row has index 98:")
print(df.iloc[49:99].tail(1))

print("\nc) .loc for specific column selection:")
print(df.loc[49:99, 'petal_length'].head())

print("\nd) .iloc for column position selection:")
print("(petal_length is at position 2):")
print(df.iloc[49:99, 2].head())

print("\n" + "="*80)
print("Summary table:")
print("| Method | Selection Type  | Inclusive? | Example                     |")
print("|--------|-----------------|------------|-----------------------------|")
print("| .loc   | Label-based     | Yes        | df.loc[49:99, 'sepal_width']|")
print("| .iloc  | Position-based  | No         | df.iloc[49:99, 1]           |")
```

#**Problem 6**: Creating a diagram

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
import pandas as pd

# Load and prepare data
iris = load_iris()
df = pd.DataFrame(iris.data, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
df['species'] = iris.target
species_names = ['setosa', 'versicolor', 'virginica']

# Set style
sns.set(style="whitegrid")

# 1. Pie Chart (in separate figure)
plt.figure(figsize=(6, 6))
species_counts = df['species'].value_counts()
plt.pie(species_counts, labels=species_names, autopct='%1.1f%%',
        colors=sns.color_palette('pastel'), startangle=90)
plt.title('Distribution of Iris Species')
plt.show()

# 2. Box Plots (in one figure)
plt.figure(figsize=(15, 10))
features = df.columns[:-1]
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x='species', y=feature, data=df, hue='species', palette='pastel', legend=False)
    plt.title(f'Box Plot of {feature.replace("_", " ").title()}')
    plt.xticks([0, 1, 2], species_names)
plt.tight_layout()
plt.show()

# 3. Violin Plots (in one figure)
plt.figure(figsize=(15, 10))
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.violinplot(x='species', y=feature, data=df, hue='species', palette='pastel', inner='quartile', legend=False)
    plt.title(f'Violin Plot of {feature.replace("_", " ").title()}')
    plt.xticks([0, 1, 2], species_names)
plt.tight_layout()
plt.show()

# Explanation of Box Plots vs Violin Plots
print("""
Box Plot vs Violin Plot Comparison:

Box Plots:
- Show quartiles (25%, 50%, 75%) and outliers
- Simple rectangular representation
- Good for comparing distributions across groups
- Clearly shows median and IQR
- Better for small datasets

Violin Plots:
- Show the full distribution of the data
- Width represents density/probability
- Combines box plot with kernel density estimation
- Reveals bimodality or unusual distributions
- Better for larger datasets

Use box plots when you need to focus on quartiles and outliers.
Use violin plots when you want to understand the complete distribution shape.
""")
```

#**Problem 7**:  Confirming the relationship between features

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
import pandas as pd

# Load and prepare data
iris = load_iris()
df = pd.DataFrame(iris.data, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
df['species'] = iris.target
species_names = ['setosa', 'versicolor', 'virginica']

# Set style
sns.set(style="whitegrid", palette="pastel")

## Individual Scatter Plots for all feature pairs (6 combinations)
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
plt.figure(figsize=(18, 12))

plot_number = 1
for i, feat1 in enumerate(features):
    for j, feat2 in enumerate(features):
        if i < j:
            plt.subplot(2, 3, plot_number)
            sns.scatterplot(x=feat1, y=feat2, hue='species', data=df, palette='viridis')
            plt.title(f'{feat1} vs {feat2}'.replace('_', ' ').title())
            plt.xlabel(feat1.replace('_', ' ').title())
            plt.ylabel(feat2.replace('_', ' ').title())
            plot_number += 1  # Increment plot counter

plt.tight_layout()
plt.show()

## Scatter Plot Matrix (Pairplot)
print("\nScatter Plot Matrix:")
sns.pairplot(df, hue='species', palette='viridis', height=2.5)
plt.show()

## Correlation Coefficient Matrix
print("\nCorrelation Coefficient Matrix:")
corr_matrix = df[features].corr()
print(corr_matrix)

## Heatmap of Correlation Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt=".2f",
            xticklabels=[f.replace('_', ' ').title() for f in features],
            yticklabels=[f.replace('_', ' ').title() for f in features])
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.show()
```
